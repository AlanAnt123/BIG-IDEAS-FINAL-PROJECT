```python
fire = pd.read_csv("firearmSeizures.csv", sep=None, engine='python')
fireGroup = fire.groupby("neighborhood", dropna=False, as_index=False).sum()
fireGroupSorted = fireGroup.sort_values(by="total_count", ascending=True).reset_index(drop=True)
fireGroupSorted = fireGroupSorted.iloc[:-1]
plt.figure(figsize=(20,25))
plt.bar(fireGroupSorted['neighborhood'].astype(str), fireGroupSorted['total_count'].astype(str))
plt.xticks(rotation=90,ha='right')
plt.xlabel('Neighborhood')
plt.ylabel('Total Firearms Seized')
plt.title('Neighborhoods ranked by most to least firearms seized')
plt.tight_layout()
plt.show()
```

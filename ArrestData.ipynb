```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('e03a89dd-134a-4ee8-a2bd-62c40aeebc6f-2.csv')

# Define neighborhood column
neighborhood_col = 'INCIDENTNEIGHBORHOOD'

# Clean up missing or blank neighborhood names
df[neighborhood_col] = df[neighborhood_col].fillna('Unknown').str.strip()

# Count total crimes per neighborhood
crime_counts = df[neighborhood_col].value_counts().reset_index()
crime_counts.columns = ['Neighborhood', 'Crime_Count']

# Rank neighborhoods from safest to least safe (lowest to highest crime)
crime_counts_sorted = crime_counts.sort_values(by='Crime_Count', ascending=True).reset_index(drop=True)
crime_counts_sorted['Rank'] = crime_counts_sorted.index + 1  # add rank numbers

# This parrint top and bottom 10 neighborhoods
print("üèÜ Top 10 Safest Neighborhoods (Lowest Crime):")
print(crime_counts_sorted.head(10).to_string(index=False))

print("\n‚ö†Ô∏è Top 10 Least Safe Neighborhoods (Highest Crime):")
print(crime_counts_sorted.tail(10).sort_values(by='Crime_Count', ascending=False).to_string(index=False))




# This is the part of my code that plots all of hte points onto the bargraph.
plt.figure(figsize=(10, len(crime_counts_sorted) * 0.25))  # auto scale height
plt.barh(crime_counts_sorted['Neighborhood'], crime_counts_sorted['Crime_Count'], color='skyblue')
plt.xlabel('Total Crime Count')
plt.ylabel('Neighborhood')
plt.title('Pittsburgh Neighborhoods Ranked from Safest (Top) to Least Safe (Bottom)')
plt.gca().invert_yaxis()  # safest (lowest crime) appears at top
plt.tight_layout()
plt.show()
```

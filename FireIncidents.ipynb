```python
import csv
from collections import Counter

csv_file_path = '8d76ac6b-5ae8-4428-82a4-043130d17b02.csv'
incident_counts = Counter()

with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        neighborhood = row.get('neighborhood') or row.get('Neighborhood') or row.get('NEIGHBORHOOD')
        if neighborhood:
            incident_counts[neighborhood.strip()] += 1

print("Incident counts by neighborhood:")
for neighborhood, count in incident_counts.items():
    print(f"{neighborhood}: {count}")

least_common = sorted(incident_counts.items(), key=lambda x: x[1])[:10]

print("\nTop 10 neighborhoods with the fewest incidents:")
for rank, (neighborhood, count) in enumerate(least_common, start=1):
    print(f"{rank}. {neighborhood} - {count} incidents")
```

```python
import pandas as pd

df = pd.read_csv('8d76ac6b-5ae8-4428-82a4-043130d17b02.csv')
col = next((c for c in df.columns if c.lower() == 'neighborhood'), None)
counts = df[col].value_counts()

print("Incident counts by neighborhood:")
for name, count in counts.items():
    print(f"{name}: {count}")

least = counts.sort_values().head(10)

print("\nTop 10 neighborhoods with the fewest incidents:")
for i, (name, count) in enumerate(least.items(), 1):
    print(f"{i}. {name} - {count} incidents")
```

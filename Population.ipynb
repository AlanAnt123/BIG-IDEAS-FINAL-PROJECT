```python
import pandas as pd

df = pd.read_csv('82f29015-6905-4b1c-8300-afe9bb2231b3.csv')
col_name = next((c for c in df.columns if c.lower() == 'neighborhood'), None)
pop_col = next((c for c in df.columns if 'estimate' in c.lower()), None)

df_filtered = df[[col_name, pop_col]].dropna()

print("Population estimates by neighborhood:")
for name, pop in zip(df_filtered[col_name], df_filtered[pop_col]):
    print(f"{name}: {int(pop)}")
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load full dataset
df = pd.read_csv('e03a89dd-134a-4ee8-a2bd-62c40aeebc6f-2.csv')

# Column for neighborhoods
neighborhood_col = 'INCIDENTNEIGHBORHOOD'

# Master list of neighborhoods (all 90)
all_neighborhoods = [
    'Allegheny Center', 'Allegheny West', 'Allentown', 'Arlington / Arlington Heights',
    'Banksville', 'Beechview', 'Beltzhoover', 'Bloomfield', 'Bon Air', 'Bluff',
    'Brighton Heights', 'Brookline', 'California‑Kirkbride', 'Carrick', 'Chartiers City',
    'Chateau', 'Crafton Heights', 'Crawford‑Roberts', 'Downtown', 'Duquesne Heights',
    'East Allegheny (aka Deutschtown)', 'East Carnegie', 'East Hills', 'East Liberty',
    'Elliott', 'Esplen', 'Fairywood', 'Fineview', 'Friendship', 'Garfield', 'Glen Hazel',
    'Greenfield', 'Hays', 'Hazelwood', 'Highland Park', 'Homewood North', 'Homewood South',
    'Homewood West', 'Knoxville', 'Larimer', 'Lincoln‑Lemington‑Belmar', 'Lincoln Place',
    'Lower Lawrenceville', 'Manchester', 'Marshall‑Shadeland', 'Morningside', 'Mount Oliver',
    'Mount Washington', 'New Homestead', 'North Oakland', 'North Shore', 'Northview Heights',
    'Oakwood', 'Overbrook', 'Perry North', 'Perry South', 'Point Breeze', 'Polish Hill',
    'Regent Square', 'Ridgemont', 'Shadyside', 'Sheraden', 'South Oakland', 'South Side Flats',
    'South Side Slopes', 'Spring Garden', 'Spring Hill‑City View', 'Squirrel Hill North',
    'Squirrel Hill South', 'St. Clair (Saint Clair)', 'Stanton Heights', 'Strip District',
    'Summer Hill', 'Swisshelm Park', 'Troy Hill', 'Uptown (aka Bluff/Soho)', 'Upper Hill',
    'Upper Lawrenceville', 'West End', 'West Oakland', 'Westwood', 'Windgap',
    'Central Lawrenceville', 'Central North Side', 'Central Oakland', 'Point Breeze North',
    'Greenfield/Four Mile Run', 'Homewood (general umbrella)', 'Oakland (combined academic hub)',
    'Hill District'
]

# Count crimes per neighborhood
crime_counts = df[neighborhood_col].value_counts()

# Create DataFrame for all neighborhoods
crime_df = pd.DataFrame({'Neighborhood': all_neighborhoods})
crime_df['crime_count'] = crime_df['Neighborhood'].map(crime_counts).fillna(0)

# Sort neighborhoods by lowest crime count (best to worst)
crime_df_sorted = crime_df.sort_values(by='crime_count', ascending=True).reset_index(drop=True)

# Print top 10 best neighborhoods
print("Top 10 neighborhoods by lowest crime rates:")
print(crime_df_sorted[['Neighborhood', 'crime_count']].head(10))

# Optional: plot all neighborhoods
plt.figure(figsize=(15,8))
plt.bar(crime_df_sorted['Neighborhood'], crime_df_sorted['crime_count'], color='skyblue')
plt.xticks(rotation=90, ha='right')
plt.xlabel('Neighborhood')
plt.ylabel('Total Crime Count')
plt.title('Neighborhoods Ranked by Lowest Crime Rate (Best = Lowest Crime)')
plt.tight_layout()
plt.show()
```
